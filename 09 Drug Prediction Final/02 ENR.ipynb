### Read CSV

```python
df = pd.read_csv("Final_df.csv")
df.head()
```

### Remove Noise from dataset

```python
df['review'] = df['review'].apply(lambda x: " ".join(x.lower() for x in x.split())) # lower case conversion
df['review'] = df['review'].str.replace('[^\w\s]','') # getting rid of special characters
df['review'] = df['review'].str.replace('\d+', '') # removing numeric values from between the words
df['review'] = df['review'].apply(lambda x: x.translate(string.digits)) # removing numerical numbers
stop = stopwords.words('english')
df['review'] = df['review'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))  # removing stop words
```

```python
df['sentiment'] = df['review'].apply(lambda tweet: TextBlob(tweet).sentiment.polarity)
```

```python
scores = []
for i in df['sentiment']:
    if i>0:
        scores.append(1)
    elif i<0:
        scores.append(-1)
    else:
        scores.append(0)
```

```python
df['sentiment'] = scores
```

```python
df.head()
```

```python
reviews = df[['review']]
```

### Medical Entities Extraction

```python
enrs = []
for k, review in reviews.iterrows():
    doc = nlp(review[-1])
    l = list(doc.ents)

    if(len(l)==0):
        print("0")
        print("=========================================")
        enrs.append("Normal")
    else:
#         listToStr = ' '.join(map(str, l))
#         print(listToStr)
        print("=========================================")
        enrs.append(l[0])
```

### Save CSV File

```python
df['ENRS'] = enrs
df.to_csv("dataset.csv", index=False)
```
