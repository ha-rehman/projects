## Data Preprocessing

```python
balanced_data = pd.DataFrame()
train_data = pd.DataFrame()
test_data =  pd.DataFrame()
```

```python
train_data = pd.concat([train_data, data[data['class'] == 0][:1000]])
train_data = pd.concat([train_data, data[data['class'] == 1][:1000]])
train_data = pd.concat([train_data, data[data['class'] == 2][:1000]])
test_data = pd.concat([test_data, data[data['class'] == 0][1000:1200]])
test_data = pd.concat([test_data, data[data['class'] == 1][1000:1200]])
test_data = pd.concat([test_data, data[data['class'] == 2][1000:1200]])
```

```python
balanced_data = pd.concat([balanced_data, train_data, test_data])
```

```python
balanced_data.reset_index(inplace=True)
```

```python
balanced_data['tweet'] = balanced_data['tweet'].apply(nfx.remove_stopwords)
```

```python
x = balanced_data.tweet
y = balanced_data[['hate_speech', 'offensive_language', 'neither']]
```

#### Apply TFIDF vectorizer on the text data

```python
tfidf = TfidfVectorizer()
Xfeatures = tfidf.fit_transform(x).toarray()
```

```python
X_train,X_test,y_train,y_test = train_test_split(Xfeatures,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

# **Comparsion Between Classification Machine Learning Models**


```python
label_results = [accuracy_score(y_test_arg, Y_pred_label), precision_score(y_test_arg, Y_pred_label, average='macro'),
              recall_score(y_test_arg, Y_pred_label, average='macro'),f1_score(y_test_arg, Y_pred_label, average='macro')]

chain_results = [accuracy_score(y_test_arg, Y_pred_chain), precision_score(y_test_arg, Y_pred_chain, average='macro'),
              recall_score(y_test_arg, Y_pred_chain, average='macro'),f1_score(y_test_arg, Y_pred_chain, average='macro')]
```

```python
from tabulate import tabulate
# showing all models result
dic = {
    'Metrics':['accuracy','precision','recall','f1-score'],
    'Chain Classifier SVM' : chain_results,
    'Label Powerset SVM' : label_results,
    
}
metrics_df = pd.DataFrame(dic)

metrics_df = metrics_df.set_index('Metrics')
# displaying the DataFrame
print(tabulate(metrics_df, headers = 'keys', tablefmt = 'psql'))
```
